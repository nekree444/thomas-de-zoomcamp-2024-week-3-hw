```python
li = []
```

```python
import pandas as pd
import datetime as dt
from time import time
import fastparquet
```

```python
taxi_dtypes = {
    'VendorID': pd.Int64Dtype(),
    'store_and_fwd_flag': str, 
    'RatecodeID': pd.Int64Dtype(),
    'PULocationID': pd.Int64Dtype(),
    'DOLocationID': pd.Int64Dtype(),
    'passenger_count': pd.Int64Dtype(),
    'trip_distance': float,
    'fare_amount': float,
    'extra': float,
    'mta_tax': float,
    'tip_amount': float,
    'tolls_amount': float,
    'improvement_surcharge': float,
    'total_amount': float,
    'payment_type': pd.Int64Dtype(),
    'trip_type': float,
    'congestion_surcharge': float,
}
```

```python
parse_dates = ['lpep_pickup_datetime', 'lpep_dropoff_datetime']
```

```python
for i in range(1, 13):
    t_start = time()
    if i <= 9:
        url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2022-0{i}.parquet"
        df = pd.read_parquet(url, engine='fastparquet')
        li.append(df)
    else:
        url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2022-{i}.parquet"
        df = pd.read_parquet(url, engine='fastparquet')
        li.append(df)
    t_end = time()
    print(f'Appended another df, took {t_end - t_start} seconds')
```

```python
df = pd.concat(li)
```

```python
df
```

```python
df.columns = (df.columns
                  .str.replace(' ', '_')
                  .str.lower()
                 )
df = df[df['passenger_count'] > 0]
df = df[df['trip_distance'] > 0]
df = df[df['vendorid'] != None]
df['lpep_pickup_date'], df['lpep_dropoff_date'] = df['lpep_pickup_datetime'].dt.date, df['lpep_dropoff_datetime'].dt.date
```
